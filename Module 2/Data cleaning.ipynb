```python
##Fill missing values with mode (most common values for that column)
df["type"].fillna(value=df['type'].mode().loc[0],inplace=True)
df.head()
```

```python
##For numerical data we can fill the missing values with the mean
print(df['calories'].mean())
```

```python
df['calories'].fillna(value=df['calories'].mean(),inplace=True)
```
